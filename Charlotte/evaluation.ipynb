```python
from validation import KFoldValidator
```

```python
clean_validator = KFoldValidator('../wifi_db/clean_dataset.txt', k=10)
clean_results = clean_validator.validate()

print("Clean Data Results:\n")
print(f"Average Accuracy: {clean_results['accuracy']:.4f}\n")
print(f"Precision per Class:\n - Class 1:{clean_results['precision_per_class'][0]:.4f}\n - Class 2:{clean_results['precision_per_class'][1]:.4f}\n - Class 3:{clean_results['precision_per_class'][2]:.4f}\n - Class 4:{clean_results['precision_per_class'][3]:.4f}\n")
print(f"Recall per Class:\n - Class 1:{clean_results['recall_per_class'][0]:.4f}\n - Class 2:{clean_results['recall_per_class'][1]:.4f}\n - Class 3:{clean_results['recall_per_class'][2]:.4f}\n - Class 4:{clean_results['recall_per_class'][3]:.4f}\n")
print(f"F1 Score per Class:\n - Class 1:{clean_results['f1_score_per_class'][0]:.4f}\n - Class 2:{clean_results['f1_score_per_class'][1]:.4f}\n - Class 3:{clean_results['f1_score_per_class'][2]:.4f}\n - Class 4:{clean_results['f1_score_per_class'][3]:.4f}\n")
print(f"Confusion Matrix:\n{clean_results['confusion_matrix']}")
```

```python
model = clean_validator.best_model
# model.visualise_tree(figsize=(15, 10))
model.draw_tree()
```
